```python
movies=pd.read_csv('C:\\Users\\moksh\\Downloads\\movie.csv')
movies
```

```python
movies.set_index('movieId',inplace=True)
movie_df=movies.drop(['title'],axis=1)
movie_df['genres']=movie_df['genres'].str.lower().str.split('|')
```

```python
movie_df.genres
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
classes=movie_df.genres.explode().unique()
mlb=MultiLabelBinarizer()
```

```python
encod=pd.DataFrame(mlb.fit_transform(movie_df.genres),columns=mlb.classes_,index=movie_df.index)
X_enc=pd.concat([movie_df,encod],axis=1).drop('genres',axis=1).astype(float)
```

```python
X_enc
```

```python
def recommend(x,X,n):
  sim=cosine_similarity(x,X).reshape(-1,)
  sim=pd.Series(sim,index=X.index,name='Sim')
  
  sort=sim.sort_values(ascending=False)
  return sort.head(n)
```

```python
print(movies.iloc[3])
```

```python
sim_df=recommend(X_enc.iloc[3,:].values.reshape(1,-1),X_enc,10)
sim_df
```

```python
pd.concat([movie_df.loc[movie_df.index.isin(list(sim_df.index))],sim_df],axis=1).sort_values(by='Sim',ascending=False)
```

```python
ratings=ratings_df.groupby(by='movieId').mean().rating
ratings
movie_df['ratings']=ratings
```

```python
recommend_df=pd.concat([movie_df.loc[sim_df.index],sim_df], axis=1).sort_values(by='Sim',ascending=False)
```
